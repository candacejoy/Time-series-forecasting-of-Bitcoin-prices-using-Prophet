```python
import numpy as np
import pandas as pd
#np.random.seed(42)
data = np.random.rand(4, 4)
df = pd.DataFrame(data, columns=['First', 'Second', 'Third', 'Fourth'])
print(df)
```

```python
df = df.rename(columns={'First': 'Random value 1', 'Second': 'Random value 2' , 'Third': 'Random value 3' , 'Fourth': 'Random value 4'})
print(df)
```

```python
descriptive_statistics = df.describe()
print(descriptive_statistics)
```

```python
null_val = df.isnull()
col_data_types = df.dtypes
print(null_val)
print(col_data_types)
```

```python
loc_method = df.loc[:, ['Random value 2', 'Random value 3']]
print(loc_method)
```

```python
index_loc_method = df.iloc[:, [1, 2]]
print(index_loc_method)
```
